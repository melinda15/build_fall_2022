```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt

titanic_data = pd.read_csv("../Pandas-Data-Manipulation-MelindaYee/titanic.csv")
titanic_data.head(10)
```

```python
# Step 1: Create a heatmap showing the correlation between the following features: Survived, Pclass, Sex, Age, and Fare 
data_for_heatmap = titanic_data[["Survived", "Pclass", "Age", "Fare"]]

corr_heatmap = seaborn.heatmap(data_for_heatmap.corr(), annot = True, cmap = "mako")
```

```python
# Step 2: Create a bar graph that illustrates how many passengers fell into each of these age ranges: 0 - 16, 17 - 25, 26 -40, 41 - 59, 60 or older
age_groups = {
    "groups": ["16 or younger", "17 to 25", "26 to 40", "41 to 59", "60+"],
    "count": [
        len(titanic_data[(titanic_data["Age"] >= 0) & (titanic_data["Age"] <= 16)]),
        len(titanic_data[(titanic_data["Age"] >=  17) & (titanic_data["Age"] <= 25)]),
        len(titanic_data[(titanic_data["Age"] >=  26) & (titanic_data["Age"] <= 40)]),
        len(titanic_data[(titanic_data["Age"] >=  41) & (titanic_data["Age"] <= 59)]),
        len(titanic_data[titanic_data["Age"] >= 60]),
    ],
}


df = pd.DataFrame(age_groups)
age_group_chart = seaborn.barplot(data = df, x = "groups", y = "count")
```

```python
# Step 3: Create a line graph showing the average survival percentage of each of the previous age groups

total_children = len(titanic_data[(titanic_data["Age"] >= 0) & (titanic_data["Age"] <= 16)])
children_survived = titanic_data[(titanic_data["Age"] >= 0) & (titanic_data["Age"] <= 16) & (titanic_data["Survived"] == 1)]["Survived"].sum()
children_survival_percent = children_survived / total_children * 100 

total_young_adults = len(titanic_data[(titanic_data["Age"] >=  17) & (titanic_data["Age"] <= 25)])
young_adults_survived = titanic_data[(titanic_data["Age"] >= 17) & (titanic_data["Age"] <= 25) & (titanic_data["Survived"] == 1)]["Survived"].sum()
young_adults_survival_percent = young_adults_survived / total_young_adults * 100 

total_adults = len(titanic_data[(titanic_data["Age"] >=  26) & (titanic_data["Age"] <= 40)])
adults_survived = titanic_data[(titanic_data["Age"] >= 26) & (titanic_data["Age"] <= 40) & (titanic_data["Survived"] == 1)]["Survived"].sum()
adults_survival_percent = adults_survived / total_adults * 100 

total_mature_adults = len(titanic_data[(titanic_data["Age"] >=  41) & (titanic_data["Age"] <= 59)])
mature_adults_survived = titanic_data[(titanic_data["Age"] >= 41) & (titanic_data["Age"] <= 59) & (titanic_data["Survived"] == 1)]["Survived"].sum()
mature_adults_survival_percent = mature_adults_survived / total_mature_adults * 100 

total_seniors = len(titanic_data[(titanic_data["Age"] >=  60)])
seniors_survived = titanic_data[(titanic_data["Age"] >= 60) & (titanic_data["Survived"] == 1)]["Survived"].sum()
seniors_survival_percent = seniors_survived / total_seniors* 100 

survival_percent_groups = {
    "groups": ["Children", "Young Adults", "Adults", "Mature Adults", "Seniors"],
    "percent": [
        children_survival_percent,
        young_adults_survival_percent,
        adults_survival_percent,
        mature_adults_survival_percent,
        seniors_survival_percent,
    ],
}

df = pd.DataFrame(survival_percent_groups)

survival_age_pers = seaborn.lineplot(data = df, x = "groups", y = "percent")
```

```python
# Step 4: Create a pie chart that shows the percentage of survivors that were male and the percentage of survivors that were female
#survivors = 
total_males = len(titanic_data[titanic_data["Sex"] == "male"])
males_survived = titanic_data[(titanic_data["Sex"] == "male") & (titanic_data["Survived"] == 1)]["Survived"].sum()
males_percent = males_survived / total_males * 100

total_females = len(titanic_data[titanic_data["Sex"] == "female"])
females_survived = titanic_data[(titanic_data["Sex"] == "female") & (titanic_data["Survived"] == 1)]["Survived"].sum()
females_percent = females_survived / total_females * 100

survival_percent_groups = {
    "groups": ["Males", "Females"],
    "percent": [
        males_percent,
        females_percent,
    ],
}

df = pd.DataFrame(survival_percent_groups)

y = df["percent"]

plt.pie(y, labels = df["groups"])
plt.show()
```
